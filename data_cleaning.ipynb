```python
import pandas as pd
```

```python
df = pd.read_csv('data_analyst_jobs.csv')
```

```python
df.head()
```

```python
salary = df['Salary Estimate'].apply(lambda x: x.replace("$", ""))
split_salary = salary.apply(lambda x: [int(x.split(" ", 1)[0].split(".")[0].replace(",", "")), x.split(" ")[1]])
split_salary
```

```python
df['Salary Estimate'] = split_salary.apply(lambda x: int(x[0]))
df 
```

```python
df["Salary Classification"] = split_salary.apply(lambda x: 1 if "/hr" in x[1] else(2 if "/mo" in x[1] else 3))
df
```

```python
df["Job State"] = df['Location'].apply(lambda x: x.split(",")[1] if "," in x else "Remote")
```

```python
df["Job State"].value_counts()
```

```python
df["Age"] = df["Founded"].apply(lambda x: 2023 - x if x != -1 else -1)
```

```python
df = df.drop_duplicates(subset=None, keep='first', inplace=False)
df.tail()
df = df.reset_index(drop=True)
```

```python
import re

without_uni_chars = df['Job Description'].apply(lambda x: re.sub('[\W_]+', ' ', x))

#Pyhton
df["Python"] = without_uni_chars.apply(lambda x: 1 if "python" in x.lower() else 0)
#Spark
df["Spark"] = without_uni_chars.apply(lambda x: 1 if "spark" in x.lower() else 0)
#Excel
df["Excel"] = without_uni_chars.apply(lambda x: 1 if "excel" in x.lower() else 0)
#Tableau
df["Tableau"] = without_uni_chars.apply(lambda x: 1 if "tableau" in x.lower() else 0)
#AWS
df["AWS"] = without_uni_chars.apply(lambda x: 1 if "aws" in x.lower() else 0)
#SQL
df["SQL"] = without_uni_chars.apply(lambda x: 1 if "sql" in x.lower() else 0)

```

```python
df.to_csv("data_analyst_jobs_cleaned.csv", index=False)
```
